```python
import pandas as pd
import numpy as np
from sqlalchemy import create_engine
```

```python
# Path to data set in a variable
wh15 = "../Project_2_Team_5/Resources/2015.csv"
wh16 = "../Project_2_Team_5/Resources/2016.csv"
wh17 = "../Project_2_Team_5/Resources/2017.csv"
wh18 = "../Project_2_Team_5/Resources/2018.csv"
wh19 = "../Project_2_Team_5/Resources/2019.csv"
```

```python
# Use Pandas to read data
wh15_df = pd.read_csv(wh15)
wh15_df.head()

# Drop columns
wh15_drop = wh15_df.drop(columns=['Region', 'Standard Error', 'Generosity','Dystopia Residual', 'Happiness Rank', 'Family', 'Health (Life Expectancy)', 'Freedom', 'Trust (Government Corruption)'])
wh15_drop.head()

# Sort ascending
wh15_sorted = wh15_drop.sort_values(by='Country', ascending=True)
wh15_sorted.head() 

# Add "Year" column
wh15_sorted.insert(loc=1,
          column='year',
          value=2015)
wh15_sorted.head()  

# Rename columns
wh15_rename = wh15_sorted.rename(columns={"Economy (GDP per Capita)": "gdp_per_capita", 
                                        "Country": "country",
                                        "Happiness Score": "happiness_score"}, 
                                errors="raise")
wh15_rename.head()    
```

```python
# Use Pandas to read data
wh16_df = pd.read_csv(wh16)
wh16_df.head()

# Drop columns
wh16_drop = wh16_df.drop(columns=['Region', 'Lower Confidence Interval', 'Upper Confidence Interval','Generosity','Dystopia Residual', 'Happiness Rank', 'Family', 'Health (Life Expectancy)', 'Freedom', 'Trust (Government Corruption)'])
wh16_drop.head()


# Sort ascending
wh16_sorted = wh16_drop.sort_values(by='Country', ascending=True)
wh16_sorted.head()

# Add "Year" column
wh16_sorted.insert(loc=1,
          column='year',
          value=2016)
wh16_sorted.head()  

#Rename columns
wh16_rename = wh16_sorted.rename(columns={"Economy (GDP per Capita)": "gdp_per_capita",
                                        "Country": "country",
                                        "Happiness Score": "happiness_score"}, 
                                errors="raise")
wh16_rename.head()
```

```python
# Use Pandas to read data
wh17_df = pd.read_csv(wh17)
wh17_df.head()

# Drop columns
wh17_drop = wh17_df.drop(columns=['Whisker.high', 'Whisker.low','Generosity', 'Dystopia.Residual', 'Happiness.Rank', 'Family', 'Health..Life.Expectancy.', 'Freedom', 'Trust..Government.Corruption.'])
wh17_drop.head()

# Sort ascending
wh17_sorted = wh17_drop.sort_values(by='Country', ascending=True)
wh17_sorted.head() 

# Add "Year" column
wh17_sorted.insert(loc=1,
          column='year',
          value=2017)
wh17_sorted.head()

#Rename columns
wh17_rename = wh17_sorted.rename(columns={"Happiness.Score": "happiness_score",
                                 "Economy..GDP.per.Capita.": "gdp_per_capita",
                                 "Country": "country"}, 
                                 errors="raise")
wh17_rename.head()  

```

```python
duplicate = wh17_rename[wh17_rename.duplicated(['country', 'year'])]
duplicate 
```

```python
# Use Pandas to read data
wh18_df = pd.read_csv(wh18)
wh18_df.head()

# Drop columns
wh18_drop = wh18_df.drop(columns=['Overall rank','Generosity', 'Social support', 'Healthy life expectancy', 'Freedom to make life choices', 'Perceptions of corruption'])
wh18_drop.head()


# Sort ascending
wh18_sorted = wh18_drop.sort_values(by='Country or region', ascending=True)
wh18_sorted.head()

# Add "Year" column
wh18_sorted.insert(loc=1,
          column='year',
          value=2018)
wh18_sorted.head()  

#Rename columns
wh18_rename = wh18_sorted.rename(columns={"Country or region": "country", 
                                        "Score": "happiness_score", 
                                        "GDP per capita": "gdp_per_capita"}, 
                                errors="raise")
wh18_rename.head()
```

```python
# Use Pandas to read data
wh19_df = pd.read_csv(wh19)
wh19_df.head()

# Drop columns
wh19_drop = wh19_df.drop(columns=['Overall rank', 'Generosity', 'Social support', 'Healthy life expectancy', 'Freedom to make life choices', 'Perceptions of corruption'])
wh19_drop.head()

# Sort ascending
wh19_sorted = wh19_drop.sort_values(by='Country or region', ascending=True)
wh19_sorted.head()

# Add "Year" column
wh19_sorted.insert(loc=1,
          column='year',
          value=2019)
wh19_sorted.head()  

#Rename columns
wh19_rename = wh19_sorted.rename(columns={"Country or region": "country", 
                                        "Score": "happiness_score", 
                                        "GDP per capita": "gdp_per_capita"}, 
                                errors="raise")
#formatting numbers to 2 decimal points                                
pd.options.display.float_format = '{:,.2f}'.format
wh19_rename.head()
```

```python
wh_append = wh15_rename\
    .append(wh16_rename, ignore_index=True)\
    .append(wh17_rename, ignore_index=True)\
    .append(wh18_rename, ignore_index=True)\
    .append(wh19_rename, ignore_index=True)\
    

wh_append

sorted = wh_append.sort_values(by='country', ascending = True)

sorted
```
